```python
import os
import pandas as pd
```

```python
file = os.listdir('good_pred')[1:]
```

```python
file
```

```python
df = pd.read_csv('good_pred/'+file[0])
for x in range(1,len(file)):
    new_df = pd.read_csv('good_pred/'+file[x])
    df = pd.merge(df, new_df, on='id')
```

```python
df.columns = ['id', '1', '2', '3', '4', '5', '6', '7', '8']
```

```python
df.head()
```

```python
weight = [0.51282, 0.51515, 0.52727, 0.53735, 0.58484, 0.56751, 0.58484, 0.54744]
```

```python
weight =  [x / sum(weight) for x in weight]
```

```python
weight
```

```python
for i in range(8):
    df[str(i+1)] = df[str(i+1)].apply(lambda x: x * weight[i])
```

```python
df.head()
```

```python
df['sum'] = (df['1'] + df['2']+ df['3'] + df['4'] + df['5'] + df['6'] + df['7'] + df['8'])
```

```python
df.head()
```

```python
def prediction(x):
    if x >= 0.5:
        return 1
    else:
        return 0
```

```python
df['pred'] = df['sum'].apply(prediction)
```

```python
df.head()
```

```python
pd.DataFrame({'id':df.id, 'any_spot':df.pred}, columns=['id','any_spot']).astype({'any_spot':int})
```

```python
pd.DataFrame({'id':df.id, 'any_spot':df.pred}, columns=['id','any_spot']).astype({'any_spot':int}).to_csv('preds/merge_good_preds.csv', index=False)
```
