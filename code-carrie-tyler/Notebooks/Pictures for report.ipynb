```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
sns.set()
```

```python
data = pd.read_csv('../Data/train-parking.csv')
```

```python
plt.figure(figsize=(10, 8))
plt.xticks(rotation = 90)
ax = sns.boxplot(x = 'Street', y = 'Real.Spots', data = data)
ax.set_ylim(0, 30)
```

```python
data['hour'] = data['Time'].str.extract('(\d+):\d+', expand = False).astype('int')
```

```python
plt.figure(figsize=(20, 8))
plt.xticks(rotation = 90)
ax = sns.boxplot(x = 'hour', y = 'Real.Spots', data = data)
ax.set_ylim(0, 30)
```

```python
vals = data.groupby('hour')['any_spot'].count()
plt.plot(vals.index, vals.values)
```

```python
_ = plt.hist(data.groupby(['Street', 'From'])['To'].count(), bins=20)
```

```python
_ = plt.hist(data.groupby(['Street'])['To'].count(), bins=30)
```

```python
data['date'] = pd.to_datetime(data['Date'])
data['date_str'] = data['date'].dt.strftime('%Y-%m-%d')
```

```python
myFmt = mdates.DateFormatter('%Y-%m-%d')

plt.figure(figsize=(8, 8))
plt.xticks(rotation = 90)
ax = sns.boxplot(x = 'date_str', y = 'Real.Spots', data = data.sort_values('date'))

plt.ylim(0, 50)
```
